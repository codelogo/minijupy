```python
import matplotlib.pyplot as plt

with open("/work/data/bin.ls", "r", encoding="utf-8") as fd:
    words = [w.strip() for w in fd]
lengths = [len(w) for w in words]

plt.figure(figsize=(12,5))
plt.hist(lengths, bins=range(1, max(lengths)+2), edgecolor='black', align='left')
plt.xlabel("Length of the words")
plt.ylabel("Number of words")
plt.title("Distribution of the lengths of the words")
plt.xticks(range(1, max(lengths)+1))
plt.show()
```
